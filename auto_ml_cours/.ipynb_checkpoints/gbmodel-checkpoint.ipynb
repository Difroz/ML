```python
def select_col(df):
    categorical_cols = [cname for cname in df.columns if df[cname].dtype == "object"]
    numerical_cols = [cname for cname in df.columns if df[cname].dtype in ['int64', 'float64']]
    return categorical_cols, numerical_cols
```
