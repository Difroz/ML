```python
import os
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
os.listdir()
```

```python
files = os.listdir()[0:2]
```

```python
df_1 = pd.read_csv(files[1], index_col='date', parse_dates=True)
df_2 = pd.read_csv(files[0])
```

```python
df_1.head()
```

```python
df_2.head()
```

```python
pca = PCA(n_components=10)
pca.fit(df_1)
```

```python
#ans_1
pca.explained_variance_ratio_[0:4]
```

```python
with open('ans_2_1', 'w') as f:
    f.write('4')
```

```python
#ans_2
comp = pd.DataFrame(pca.transform(df_1))
comp[0].values
```

```python
df_2['^DJI'].values
```

```python
np.corrcoef(comp[0].values, df_2['^DJI'].values)
```

```python
result = np.round(np.corrcoef(comp[0].values, df_2['^DJI'].values)[0][1], 2).astype('str')
```

```python
result
```

```python
with open('ans_2_2', 'w') as f:
    f.write(result)
    
```

```python
#ans_3
res_3 = pca.components_[0].tolist()
res_3 = res_3.index(max(res_3))
with open('ans_2_3', 'w') as f:
    f.write(df_1.columns[res_3])
```
