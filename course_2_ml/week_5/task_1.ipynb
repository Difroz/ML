```python
import os
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import KFold, cross_val_score
import pandas as pd
```

```python
os.listdir()
```

```python
df = pd.read_csv('abalone.csv')
df
```

```python
df['Sex'] = df['Sex'].map({'I':0, 'F':-1,'M':1})
```

```python
df
```

```python
y = df.iloc[:,-1]
X = df.iloc[:,:-1]
```

```python
y
```

```python
X
```

```python
kf = KFold(shuffle=True, random_state=1)
```

```python
def cross_val(X, y):
    result = {}
    for k in range(1,50):
        model = RandomForestRegressor(n_estimators=k, random_state=1)
        cross = cross_val_score(model, X, y, cv=kf, scoring='r2')
        result[k] = round(cross.mean(), 2)
    return result
```

```python
result = cross_val(X,y)
result
```
